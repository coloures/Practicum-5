```python
def is_integer_series(series):
    # Убираем NaN, проверяем, что все значения — числа
    if not pd.api.types.is_numeric_dtype(series):
        return False
    # Проверяем, что каждое не-NaN значение равно своему целому аналогу
    return series.dropna().apply(lambda x: x == int(x)).all()
```
